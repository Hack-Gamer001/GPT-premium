```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
import os
import cv2
from tensorflow.keras.callbacks import EarlyStopping, ModelCheckpoint, ReduceLROnPlateau
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from sklearn.metrics import classification_report, confusion_matrix
import seaborn as sns

# Configuraciones generales
IMAGEN_TAMANO = 64  # Tiny ImageNet tiene imágenes de 64x64
BATCH_SIZE = 64
EPOCHS = 10
NUM_CLASES = 200

# 1. Función para cargar el dataset Tiny ImageNet
def cargar_tiny_imagenet(ruta_base):
    """
    Carga el dataset Tiny ImageNet desde una ruta base.
    Estructura esperada:
    - ruta_base/train/ (clases en subdirectorios)
    - ruta_base/val/ (clases en subdirectorios)
    """
    print("Cargando dataset Tiny ImageNet...")

    # Crear generadores de datos con aumento de datos para entrenamiento
    train_datagen = ImageDataGenerator(
        rescale=1./255,
        rotation_range=20,
        width_shift_range=0.2,
        height_shift_range=0.2,
        horizontal_flip=True,
        zoom_range=0.2,
        fill_mode='nearest',
        validation_split=0.1  # Separar 10% para validación
    )

    # Generador para datos de entrenamiento
    train_generator = train_datagen.flow_from_directory(
        os.path.join(ruta_base, 'train'),
        target_size=(IMAGEN_TAMANO, IMAGEN_TAMANO),
        batch_size=BATCH_SIZE,
        class_mode='categorical',
        subset='training'
    )

    # Generador para datos de validación del conjunto de entrenamiento
    validation_generator = train_datagen.flow_from_directory(
        os.path.join(ruta_base, 'train'),
        target_size=(IMAGEN_TAMANO, IMAGEN_TAMANO),
        batch_size=BATCH_SIZE,
        class_mode='categorical',
        subset='validation'
    )

    # Generador para datos de prueba (usando el directorio val)
    test_datagen = ImageDataGenerator(rescale=1./255)
    test_generator = test_datagen.flow_from_directory(
        os.path.join(ruta_base, 'val'),
        target_size=(IMAGEN_TAMANO, IMAGEN_TAMANO),
        batch_size=BATCH_SIZE,
        class_mode='categorical',
        shuffle=False
    )

    return train_generator, validation_generator, test_generator

# 2. Visualización de algunas imágenes del dataset
def visualizar_imagenes(generator, num_imagenes=25):
    images, labels = next(generator)
    class_names = list(generator.class_indices.keys())

    # Calcular el número de filas y columnas para la cuadrícula
    n_cols = 5
    n_rows = (num_imagenes + n_cols - 1) // n_cols

    plt.figure(figsize=(15, 15))
    for i in range(min(num_imagenes, len(images))):
        plt.subplot(n_rows, n_cols, i+1)
        # Convertir la imagen de escala [0,1] a [0,255]
        img = images[i] * 255
        img = img.astype(np.uint8)
        plt.imshow(img)
        class_idx = np.argmax(labels[i])
        plt.title(f"Clase: {class_names[class_idx]}")
        plt.axis('off')
    plt.tight_layout()
    plt.show()

# 3. Definición del modelo CNN para Tiny ImageNet
def crear_modelo_cnn():
    modelo = tf.keras.models.Sequential([
        # Primera capa convolucional
        tf.keras.layers.Conv2D(64, (3, 3), padding='same', activation='relu', input_shape=(IMAGEN_TAMANO, IMAGEN_TAMANO, 3)),
        tf.keras.layers.BatchNormalization(),
        tf.keras.layers.Conv2D(64, (3, 3), padding='same', activation='relu'),
        tf.keras.layers.BatchNormalization(),
        tf.keras.layers.MaxPooling2D(pool_size=(2, 2)),
        tf.keras.layers.Dropout(0.2),

        # Segunda capa convolucional
        tf.keras.layers.Conv2D(128, (3, 3), padding='same', activation='relu'),
        tf.keras.layers.BatchNormalization(),
        tf.keras.layers.Conv2D(128, (3, 3), padding='same', activation='relu'),
        tf.keras.layers.BatchNormalization(),
        tf.keras.layers.MaxPooling2D(pool_size=(2, 2)),
        tf.keras.layers.Dropout(0.3),

        # Tercera capa convolucional
        tf.keras.layers.Conv2D(256, (3, 3), padding='same', activation='relu'),
        tf.keras.layers.BatchNormalization(),
        tf.keras.layers.Conv2D(256, (3, 3), padding='same', activation='relu'),
        tf.keras.layers.BatchNormalization(),
        tf.keras.layers.MaxPooling2D(pool_size=(2, 2)),
        tf.keras.layers.Dropout(0.4),

        # Cuarta capa convolucional
        tf.keras.layers.Conv2D(512, (3, 3), padding='same', activation='relu'),
        tf.keras.layers.BatchNormalization(),
        tf.keras.layers.Conv2D(512, (3, 3), padding='same', activation='relu'),
        tf.keras.layers.BatchNormalization(),
        tf.keras.layers.MaxPooling2D(pool_size=(2, 2)),
        tf.keras.layers.Dropout(0.5),

        # Aplanar y capas densas
        tf.keras.layers.Flatten(),
        tf.keras.layers.Dense(1024, activation='relu'),
        tf.keras.layers.BatchNormalization(),
        tf.keras.layers.Dropout(0.5),
        tf.keras.layers.Dense(NUM_CLASES, activation='softmax')
    ])

    modelo.compile(
        optimizer=tf.keras.optimizers.Adam(learning_rate=0.001),
        loss='categorical_crossentropy',
        metrics=['accuracy', tf.keras.metrics.TopKCategoricalAccuracy(k=5, name='top5_acc')]
    )

    return modelo

# 4. Definición de modelo basado en transfer learning
def crear_modelo_transfer_learning():
    # Usar MobileNetV2 preentrenado con ImageNet
    base_model = tf.keras.applications.MobileNetV2(
        input_shape=(IMAGEN_TAMANO, IMAGEN_TAMANO, 3),
        include_top=False,
        weights='imagenet'
    )

    # Congelar el modelo base
    base_model.trainable = False

    # Crear nuevo modelo sobre la base
    modelo = tf.keras.Sequential([
        base_model,
        tf.keras.layers.GlobalAveragePooling2D(),
        tf.keras.layers.Dense(1024, activation='relu'),
        tf.keras.layers.BatchNormalization(),
        tf.keras.layers.Dropout(0.5),
        tf.keras.layers.Dense(NUM_CLASES, activation='softmax')
    ])

    modelo.compile(
        optimizer=tf.keras.optimizers.Adam(learning_rate=0.001),
        loss='categorical_crossentropy',
        metrics=['accuracy', tf.keras.metrics.TopKCategoricalAccuracy(k=5, name='top5_acc')]
    )

    return modelo

# 5. Entrenamiento del modelo
def entrenar_modelo(modelo, train_generator, validation_generator):
    # Configurar callbacks para mejorar el entrenamiento
    callbacks = [
        EarlyStopping(
            monitor='val_accuracy',
            patience=10,
            restore_best_weights=True
        ),
        ModelCheckpoint(
            filepath='mejor_modelo_tiny_imagenet.h5',
            monitor='val_accuracy',
            save_best_only=True
        ),
        ReduceLROnPlateau(
            monitor='val_loss',
            factor=0.2,
            patience=5,
            min_lr=0.00001
        )
    ]

    # Calcular pasos por época
    pasos_entrenamiento = train_generator.samples // BATCH_SIZE
    pasos_validacion = validation_generator.samples // BATCH_SIZE

    # Entrenar el modelo
    history = modelo.fit(
        train_generator,
        steps_per_epoch=pasos_entrenamiento,
        epochs=EPOCHS,
        validation_data=validation_generator,
        validation_steps=pasos_validacion,
        callbacks=callbacks
    )

    return history

# 6. Evaluación del modelo
def evaluar_modelo(modelo, test_generator):
    # Evaluar modelo
    resultados = modelo.evaluate(test_generator)
    print(f"Pérdida en test: {resultados[0]:.4f}")
    print(f"Precisión en test: {resultados[1]:.4f}")
    print(f"Top-5 Precisión en test: {resultados[2]:.4f}")

    # Hacer predicciones en todo el conjunto de prueba
    test_generator.reset()
    predictions = modelo.predict(test_generator, steps=np.ceil(test_generator.samples/BATCH_SIZE))
    y_pred = np.argmax(predictions, axis=1)

    # Obtener las etiquetas reales
    y_true = test_generator.classes

    # Mostrar reporte de clasificación (primeras 10 clases por claridad)
    class_names = list(test_generator.class_indices.keys())
    print("\nReporte de clasificación (primeras 10 clases):")
    print(classification_report(
        y_true,
        y_pred,
        target_names=class_names[:10],
        zero_division=0
    ))

    # Matriz de confusión (usar un subset para visualización)
    plt.figure(figsize=(12, 10))
    subset_size = 10  # Mostrar solo las primeras 10 clases para claridad
    cm = confusion_matrix(y_true, y_pred)[:subset_size, :subset_size]
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
                xticklabels=class_names[:subset_size],
                yticklabels=class_names[:subset_size])
    plt.xlabel('Predicción')
    plt.ylabel('Real')
    plt.title('Matriz de Confusión (primeras 10 clases)')
    plt.tight_layout()
    plt.show()

# 7. Visualizar predicciones
def visualizar_predicciones(modelo, test_generator, num_imagenes=10):
    test_generator.reset()
    images, labels = next(test_generator)
    predictions = modelo.predict(images)

    class_names = list(test_generator.class_indices.keys())

    plt.figure(figsize=(20, 10))
    for i in range(num_imagenes):
        plt.subplot(2, 5, i+1)
        plt.imshow(images[i])

        true_class_idx = np.argmax(labels[i])
        pred_class_idx = np.argmax(predictions[i])

        if true_class_idx == pred_class_idx:
            color = 'green'
        else:
            color = 'red'

        plt.title(f"Real: {class_names[true_class_idx]}\nPred: {class_names[pred_class_idx]}",
                  color=color, fontsize=10)
        plt.axis('off')

    plt.tight_layout()
    plt.show()

# 8. Visualizar el entrenamiento
def visualizar_entrenamiento(history):
    plt.figure(figsize=(15, 5))

    # Gráfico de precisión
    plt.subplot(1, 2, 1)
    plt.plot(history.history['accuracy'], label='Entrenamiento')
    plt.plot(history.history['val_accuracy'], label='Validación')
    plt.title('Precisión del modelo')
    plt.xlabel('Épocas')
    plt.ylabel('Precisión')
    plt.legend()

    # Gráfico de pérdida
    plt.subplot(1, 2, 2)
    plt.plot(history.history['loss'], label='Entrenamiento')
    plt.plot(history.history['val_loss'], label='Validación')
    plt.title('Pérdida del modelo')
    plt.xlabel('Épocas')
    plt.ylabel('Pérdida')
    plt.legend()

    plt.tight_layout()
    plt.show()

# 9. Fine-tuning del modelo de transfer learning
def fine_tuning(modelo):
    # Descongelar algunas capas del modelo base
    base_model = modelo.layers[0]
    base_model.trainable = True

    # Congelar las primeras capas y entrenar solo las últimas
    for layer in base_model.layers[:-30]:
        layer.trainable = False

    # Recompilar el modelo con una tasa de aprendizaje menor
    modelo.compile(
        optimizer=tf.keras.optimizers.Adam(learning_rate=0.0001),
        loss='categorical_crossentropy',
        metrics=['accuracy', tf.keras.metrics.TopKCategoricalAccuracy(k=5, name='top5_acc')]
    )

    return modelo

# 10. Función principal
def main():
    # Descargar el dataset Tiny ImageNet si es necesario
    print("Descargando y preparando Tiny ImageNet...")
    # Aquí debería ir el código para descargar el dataset si no está presente
    # Como ejemplo, usamos una ruta local donde se supone que está el dataset
    ruta_tiny_imagenet = "tiny-imagenet-200"

    # Cargar dataset
    train_generator, validation_generator, test_generator = cargar_tiny_imagenet(ruta_tiny_imagenet)

    # Visualizar algunas imágenes
    print("Visualizando ejemplos del dataset:")
    visualizar_imagenes(train_generator)

    # Crear modelo (elegir entre CNN estándar o transfer learning)
    modelo_tipo = "transfer"  # Opciones: "cnn" o "transfer"

    if modelo_tipo == "cnn":
        print("Creando modelo CNN personalizado...")
        modelo = crear_modelo_cnn()
    else:
        print("Creando modelo con transfer learning (MobileNetV2)...")
        modelo = crear_modelo_transfer_learning()

    # Mostrar resumen del modelo
    modelo.summary()

    # Entrenar modelo
    print("Entrenando modelo...")
    history = entrenar_modelo(modelo, train_generator, validation_generator)

    # Visualizar entrenamiento
    visualizar_entrenamiento(history)

    # Evaluar modelo
    print("Evaluando modelo...")
    evaluar_modelo(modelo, test_generator)

    # Visualizar predicciones
    print("Visualizando predicciones...")
    visualizar_predicciones(modelo, test_generator)

    # Para transfer learning, aplicar fine-tuning
    if modelo_tipo == "transfer":
        print("Aplicando fine-tuning al modelo...")
        modelo = fine_tuning(modelo)

        # Entrenar con fine-tuning
        print("Entrenando modelo con fine-tuning...")
        history_ft = entrenar_modelo(modelo, train_generator, validation_generator)

        # Visualizar entrenamiento con fine-tuning
        visualizar_entrenamiento(history_ft)

        # Evaluar modelo con fine-tuning
        print("Evaluando modelo con fine-tuning...")
        evaluar_modelo(modelo, test_generator)

        # Visualizar predicciones con fine-tuning
        print("Visualizando predicciones con fine-tuning...")
        visualizar_predicciones(modelo, test_generator)

# Descarga y preparación del dataset Tiny ImageNet
def descargar_preparar_tiny_imagenet():
    """
    Esta función descarga el dataset Tiny ImageNet y lo prepara
    en la estructura adecuada.
    """
    import os
    import urllib.request
    import tarfile
    import shutil
    from tqdm import tqdm

    # URL del dataset Tiny ImageNet
    url = "http://cs231n.stanford.edu/tiny-imagenet-200.zip"
    destino = "tiny-imagenet-200.zip"

    # Si el directorio ya existe, asumimos que ya está descargado
    if os.path.exists("tiny-imagenet-200"):
        print("El dataset ya existe en el directorio.")
        return

    # Descargar el dataset
    print(f"Descargando Tiny ImageNet desde {url}...")

    try:
        # Descargar con barra de progreso
        class DownloadProgressBar(tqdm):
            def update_to(self, b=1, bsize=1, tsize=None):
                if tsize is not None:
                    self.total = tsize
                self.update(b * bsize - self.n)

        with DownloadProgressBar(unit='B', unit_scale=True, miniters=1, desc="Tiny ImageNet") as t:
            urllib.request.urlretrieve(url, destino, reporthook=t.update_to)

        # Extraer el archivo zip
        print("Extrayendo archivos...")
        import zipfile
        with zipfile.ZipFile(destino, 'r') as zip_ref:
            zip_ref.extractall(".")

        # Reestructurar el directorio de validación para que tenga el mismo formato que el de entrenamiento
        # (subdirectorios por clase)
        print("Reestructurando directorio de validación...")
        val_dir = "tiny-imagenet-200/val"
        val_annot_file = os.path.join(val_dir, "val_annotations.txt")

        # Leer el archivo de anotaciones
        with open(val_annot_file, 'r') as f:
            val_annot = f.readlines()

        # Crear directorios por clase
        for line in val_annot:
            parts = line.strip().split('\t')
            img_name, class_id = parts[0], parts[1]

            # Crear directorio si no existe
            class_dir = os.path.join(val_dir, class_id)
            if not os.path.exists(class_dir):
                os.makedirs(class_dir)

            # Mover imagen al directorio de su clase
            img_path = os.path.join(val_dir, "images", img_name)
            if os.path.exists(img_path):
                shutil.move(img_path, os.path.join(class_dir, img_name))

        # Limpiar archivos y directorios no necesarios
        os.remove(val_annot_file)
        if os.path.exists(os.path.join(val_dir, "images")):
            shutil.rmtree(os.path.join(val_dir, "images"))

        # Eliminar archivo zip para ahorrar espacio
        os.remove(destino)

        print("Dataset preparado con éxito.")

    except Exception as e:
        print(f"Error al descargar o preparar el dataset: {e}")

# Si se ejecuta como script principal
if __name__ == "__main__":
    # Descargar y preparar el dataset
    descargar_preparar_tiny_imagenet()

    # Ejecutar el programa principal
    main()
```
