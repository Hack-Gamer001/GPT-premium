```python
# Modelo RNN Simple para Traductor Español-Quechua
# Para Google Colab - Versión mejorada

import numpy as np
import tensorflow as tf
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, SimpleRNN, Dense, RepeatVector, TimeDistributed, Dropout
import re
import unicodedata
import matplotlib.pyplot as plt
import random
import spacy
import pandas as pd
from sklearn.metrics import classification_report, confusion_matrix
import seaborn as sns

# Cargar modelo de Spacy para español
nlp = spacy.load('es_core_news_sm')

# Constantes y configuración
MAX_LENGTH_SPANISH = 40
MAX_LENGTH_QUECHUA = 40
EMBEDDING_DIM = 128
BATCH_SIZE = 64
EPOCHS = 15

# Función para normalizar texto
def unicode_to_ascii(s):
    """Normaliza caracteres unicode a ASCII"""
    return ''.join(c for c in unicodedata.normalize('NFD', s)
                  if unicodedata.category(c) != 'Mn')

def preprocess_sentence(sentence):
    """Preprocesa una oración"""
    # Convertir a ASCII
    sentence = unicode_to_ascii(sentence.lower().strip())

    # Añadir espacios alrededor de los signos de puntuación
    sentence = re.sub(r'([!.?,])', r' \1 ', sentence)

    # Eliminar caracteres no deseados
    sentence = re.sub(r'[^a-zA-ZñÑáéíóúÁÉÍÓÚüÜ.!?,]', ' ', sentence)

    # Eliminar espacios múltiples
    sentence = re.sub(r'\s+', ' ', sentence)

    return sentence.strip()

# Función para segmentar frases y realizar preprocesamiento adicional
def preprocess_and_segment(sentences):
    segmented_sentences = []
    pos_tags = []
    lemmas = []
    for sentence in sentences:
        doc = nlp(sentence)
        lemmatized_sentence = ' '.join([token.lemma_ for token in doc])
        pos_tag = ' '.join([token.pos_ for token in doc])
        segmented_sentences.append(lemmatized_sentence)
        pos_tags.append(pos_tag)
        lemmas.append([token.lemma_ for token in doc])
    return segmented_sentences, pos_tags, lemmas

# Crear dataset de muestra
def create_sample_dataset():
    """Crea un conjunto de datos de muestra para pruebas"""
    # Frases básicas en español y quechua
    data = [
        ("hola", "napaykullayki"),
        ("buenos días", "allin p'unchay"),
        ("¿cómo estás?", "imaynallan kashanki"),
        ("estoy bien", "allillanmi kachkani"),
        ("gracias", "añay"),
        ("me llamo Juan", "noqap sutiy Juan"),
        ("¿cuál es tu nombre?", "ima sutiykin qan"),
        ("adiós", "tupananchiskama"),
        ("hasta mañana", "paqarinkama"),
        ("¿dónde está la plaza?", "maypitaq plaza kashan"),
        ("necesito ayuda", "yanapayta munani"),
        ("tengo hambre", "yarqasqa kachkani"),
        ("tengo sed", "ch'akisqa kachkani"),
        ("me gusta la música", "takiyta munani"),
        ("hablas quechua", "runasimita rimanki"),
        ("estoy aprendiendo quechua", "runasimita yachachkani"),
        ("eres mi amigo", "qanmi kanki amigoy"),
        ("el cielo es azul", "hanaq pachaqa anqasmi"),
        ("la montaña es grande", "orqoqa hatunmi"),
        ("el agua es vida", "unuqa kawsaymi"),
        ("el sol brilla", "intiqa k'anchan"),
        ("la luna es hermosa", "killaqa sumaqmi"),
        ("me gusta cantar", "takiyta munani"),
        ("quiero bailar", "tusuy munani"),
        ("estoy feliz", "kusisqa kachkani"),
        ("estoy triste", "llakisqa kachkani"),
        ("te quiero", "munakuyki"),
        ("te amo", "anchata munakuyki"),
        ("es importante", "chaniyuqmi"),
        ("es necesario", "necesariom"),
    ]

    # Crear múltiples oraciones combinando las frases básicas
    spanish = []
    quechua = []

    # Agregar frases básicas
    for es, qu in data:
        spanish.append(es)
        quechua.append(qu)

    # Combinar algunas frases para crear oraciones más complejas
    for _ in range(20):
        idx1 = random.randint(0, len(data)-1)
        idx2 = random.randint(0, len(data)-1)
        spanish.append(f"{data[idx1][0]} y {data[idx2][0]}")
        quechua.append(f"{data[idx1][1]} {data[idx2][1]}")

    return spanish, quechua

# Preprocesar datos
def prepare_data():
    # Obtener datos de muestra
    spanish_sentences, quechua_sentences = create_sample_dataset()

    # Preprocesar las oraciones
    spanish_sentences, pos_tags, lemmas = preprocess_and_segment(spanish_sentences)
    quechua_sentences = [preprocess_sentence(s) for s in quechua_sentences]

    # Mostrar tabla de preprocesamiento
    preprocessing_df = pd.DataFrame({
        'Original': spanish_sentences,
        'Lemmatized': lemmas,
        'POS Tags': pos_tags
    })
    print("Tabla de Preprocesamiento:")
    print(preprocessing_df)

    # Añadir tokens especiales
    spanish_sentences = ['<start> ' + s + ' <end>' for s in spanish_sentences]
    quechua_sentences = ['<start> ' + s + ' <end>' for s in quechua_sentences]

    # Crear tokenizadores
    spanish_tokenizer = Tokenizer(filters='')
    quechua_tokenizer = Tokenizer(filters='')

    # Ajustar los tokenizadores
    spanish_tokenizer.fit_on_texts(spanish_sentences)
    quechua_tokenizer.fit_on_texts(quechua_sentences)

    # Convertir texto a secuencias
    spanish_sequences = spanish_tokenizer.texts_to_sequences(spanish_sentences)
    quechua_sequences = quechua_tokenizer.texts_to_sequences(quechua_sentences)

    # Padding de secuencias
    spanish_padded = pad_sequences(spanish_sequences, maxlen=MAX_LENGTH_SPANISH, padding='post')
    quechua_padded = pad_sequences(quechua_sequences, maxlen=MAX_LENGTH_QUECHUA, padding='post')

    # Tamaño de los vocabularios
    spanish_vocab_size = len(spanish_tokenizer.word_index) + 1  # +1 para el token 0 (padding)
    quechua_vocab_size = len(quechua_tokenizer.word_index) + 1  # +1 para el token 0 (padding)

    # Dividir en conjuntos de entrenamiento y prueba (80% - 20%)
    indices = np.arange(len(spanish_padded))
    np.random.shuffle(indices)

    train_size = int(0.8 * len(indices))

    train_indices = indices[:train_size]
    test_indices = indices[train_size:]

    x_train = spanish_padded[train_indices]
    y_train = quechua_padded[train_indices]
    x_test = spanish_padded[test_indices]
    y_test = quechua_padded[test_indices]

    # Preparar y_train y y_test para entrenamiento (one-hot encoding)
    y_train_oh = tf.keras.utils.to_categorical(y_train, num_classes=quechua_vocab_size)
    y_test_oh = tf.keras.utils.to_categorical(y_test, num_classes=quechua_vocab_size)

    return (spanish_tokenizer, quechua_tokenizer, x_train, y_train, y_train_oh,
            x_test, y_test, y_test_oh, spanish_vocab_size, quechua_vocab_size)

# Construir el modelo RNN simple
def build_rnn_model(spanish_vocab_size, quechua_vocab_size):
    model = Sequential([
        # Capa de embedding para convertir índices de palabras en vectores densos
        Embedding(input_dim=spanish_vocab_size, output_dim=EMBEDDING_DIM, mask_zero=True),

        # Capa RNN simple para codificar la secuencia de entrada
        SimpleRNN(256, return_sequences=False),

        # Capa de dropout para regularización
        Dropout(0.2),

        # Repetir el vector para la secuencia de salida
        RepeatVector(MAX_LENGTH_QUECHUA),

        # RNN decodificador que genera la secuencia de salida
        SimpleRNN(256, return_sequences=True),

        # Capa de dropout
        Dropout(0.2),

        # Capa de salida (una distribución de probabilidad sobre el vocabulario quechua para cada posición)
        TimeDistributed(Dense(quechua_vocab_size, activation='softmax'))
    ])

    # Compilar el modelo
    model.compile(
        optimizer='adam',
        loss='categorical_crossentropy',
        metrics=['accuracy']
    )

    return model

# Función para traducir una oración
def translate_sentence(sentence, model, spanish_tokenizer, quechua_tokenizer):
    # Preprocesar la oración
    processed_sentence = preprocess_sentence(sentence)
    processed_sentence = '<start> ' + processed_sentence + ' <end>'

    # Convertir a secuencia
    sequence = spanish_tokenizer.texts_to_sequences([processed_sentence])

    # Padding
    padded_sequence = pad_sequences(sequence, maxlen=MAX_LENGTH_SPANISH, padding='post')

    # Predecir con el modelo
    output_seq = model.predict(padded_sequence)

    # Convertir las probabilidades en índices de tokens
    output_seq = np.argmax(output_seq, axis=-1)

    # Convertir índices a palabras
    translated_words = []
    for idx in output_seq[0]:
        if idx > 0:  # Ignorar padding (0)
            word = get_word_from_index(idx, quechua_tokenizer)
            if word == '<end>':
                break
            if word != '<start>':
                translated_words.append(word)

    return ' '.join(translated_words)

def get_word_from_index(idx, tokenizer):
    """Obtiene una palabra a partir de su índice en el tokenizador"""
    for word, index in tokenizer.word_index.items():
        if index == idx:
            return word
    return ""

# Función para mostrar gráficos de entrenamiento
def plot_training_history(history):
    plt.figure(figsize=(12, 5))

    plt.subplot(1, 2, 1)
    plt.plot(history.history['loss'], label='Training Loss')
    plt.plot(history.history['val_loss'], label='Validation Loss')
    plt.title('Loss durante el entrenamiento')
    plt.xlabel('Época')
    plt.ylabel('Loss')
    plt.legend()

    plt.subplot(1, 2, 2)
    plt.plot(history.history['accuracy'], label='Training Accuracy')
    plt.plot(history.history['val_accuracy'], label='Validation Accuracy')
    plt.title('Precisión durante el entrenamiento')
    plt.xlabel('Época')
    plt.ylabel('Precisión')
    plt.legend()

    plt.tight_layout()
    plt.show()

# Función principal
def main():
    print("Iniciando proyecto de traducción español-quechua con RNN simple")

    # Preparar datos
    (spanish_tokenizer, quechua_tokenizer, x_train, y_train, y_train_oh,
     x_test, y_test, y_test_oh, spanish_vocab_size, quechua_vocab_size) = prepare_data()

    print(f"Tamaño del vocabulario español: {spanish_vocab_size}")
    print(f"Tamaño del vocabulario quechua: {quechua_vocab_size}")
    print(f"Ejemplos de entrenamiento: {len(x_train)}")
    print(f"Ejemplos de prueba: {len(x_test)}")

    # Construir modelo RNN
    print("\n=== Construyendo modelo RNN Simple ===")
    rnn_model = build_rnn_model(spanish_vocab_size, quechua_vocab_size)

    # Forzar la construcción del modelo con una entrada ficticia
    rnn_model(np.zeros((1, MAX_LENGTH_SPANISH)))
    rnn_model.summary()

    # Entrenar el modelo RNN
    print("\n=== Entrenando modelo RNN Simple ===")
    history = rnn_model.fit(
        x_train, y_train_oh,
        batch_size=BATCH_SIZE,
        epochs=EPOCHS,
        validation_split=0.2
    )

    # Mostrar gráficos de entrenamiento
    plot_training_history(history)

    # Evaluar modelo RNN
    loss, acc = rnn_model.evaluate(x_test, y_test_oh)
    print(f"RNN - Loss en prueba: {loss:.4f}")
    print(f"RNN - Precisión en prueba: {acc:.4f}")

    # Traducir algunas oraciones de prueba
    print("\n=== Ejemplos de traducción con RNN ===")
    test_sentences = [
        "hola amigo",
        "buenos días, ¿cómo estás?",
        "me gusta la montaña",
        "estoy aprendiendo quechua",
        "el agua es vida"
    ]

    for sentence in test_sentences:
        translation = translate_sentence(sentence, rnn_model, spanish_tokenizer, quechua_tokenizer)
        print(f"ES: {sentence}")
        print(f"QU: {translation}")
        print()

    # Guardar modelo
    rnn_model.save('rnn_esp_quechua_translator.h5')
    print("\nModelo RNN guardado con éxito.")

    print("\n=== Análisis del modelo RNN ===")
    print("Ventajas:")
    print("- Estructura más simple con menos parámetros")
    print("- Entrenamiento más rápido")
    print("- Buen rendimiento para secuencias cortas")
    print("- Menor uso de memoria")

    print("\nDesventajas:")
    print("- Problema de desvanecimiento del gradiente en secuencias largas")
    print("- Dificultad para capturar dependencias a largo plazo")
    print("- Menos preciso que LSTM en traducciones complejas")
    print("- Puede olvidar información al inicio de la secuencia cuando procesa oraciones largas")

    print("\nProyecto RNN completado con éxito!")

if __name__ == "__main__":
    main()
```

```python
# Modelo LSTM para Traductor Español-Quechua
# Para Google Colab - Versión mejorada

import numpy as np
import tensorflow as tf
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, LSTM, Dense, RepeatVector, TimeDistributed, Dropout
import re
import unicodedata
import matplotlib.pyplot as plt
import random
import spacy
import pandas as pd
from sklearn.metrics import classification_report, confusion_matrix
import seaborn as sns

# Cargar modelo de Spacy para español
nlp = spacy.load('es_core_news_sm')

# Constantes y configuración
MAX_LENGTH_SPANISH = 40
MAX_LENGTH_QUECHUA = 40
EMBEDDING_DIM = 128
BATCH_SIZE = 64
EPOCHS = 15

# Función para normalizar texto
def unicode_to_ascii(s):
    """Normaliza caracteres unicode a ASCII"""
    return ''.join(c for c in unicodedata.normalize('NFD', s)
                  if unicodedata.category(c) != 'Mn')

def preprocess_sentence(sentence):
    """Preprocesa una oración"""
    # Convertir a ASCII
    sentence = unicode_to_ascii(sentence.lower().strip())

    # Añadir espacios alrededor de los signos de puntuación
    sentence = re.sub(r'([!.?,])', r' \1 ', sentence)

    # Eliminar caracteres no deseados
    sentence = re.sub(r'[^a-zA-ZñÑáéíóúÁÉÍÓÚüÜ.!?,]', ' ', sentence)

    # Eliminar espacios múltiples
    sentence = re.sub(r'\s+', ' ', sentence)

    return sentence.strip()

# Función para segmentar frases y realizar preprocesamiento adicional
def preprocess_and_segment(sentences):
    segmented_sentences = []
    pos_tags = []
    lemmas = []
    for sentence in sentences:
        doc = nlp(sentence)
        lemmatized_sentence = ' '.join([token.lemma_ for token in doc])
        pos_tag = ' '.join([token.pos_ for token in doc])
        segmented_sentences.append(lemmatized_sentence)
        pos_tags.append(pos_tag)
        lemmas.append([token.lemma_ for token in doc])
    return segmented_sentences, pos_tags, lemmas

# Crear dataset de muestra
def create_sample_dataset():
    """Crea un conjunto de datos de muestra para pruebas"""
    # Frases básicas en español y quechua
    data = [
        ("hola", "napaykullayki"),
        ("buenos días", "allin p'unchay"),
        ("¿cómo estás?", "imaynallan kashanki"),
        ("estoy bien", "allillanmi kachkani"),
        ("gracias", "añay"),
        ("me llamo Juan", "noqap sutiy Juan"),
        ("¿cuál es tu nombre?", "ima sutiykin qan"),
        ("adiós", "tupananchiskama"),
        ("hasta mañana", "paqarinkama"),
        ("¿dónde está la plaza?", "maypitaq plaza kashan"),
        ("necesito ayuda", "yanapayta munani"),
        ("tengo hambre", "yarqasqa kachkani"),
        ("tengo sed", "ch'akisqa kachkani"),
        ("me gusta la música", "takiyta munani"),
        ("hablas quechua", "runasimita rimanki"),
        ("estoy aprendiendo quechua", "runasimita yachachkani"),
        ("eres mi amigo", "qanmi kanki amigoy"),
        ("el cielo es azul", "hanaq pachaqa anqasmi"),
        ("la montaña es grande", "orqoqa hatunmi"),
        ("el agua es vida", "unuqa kawsaymi"),
        ("el sol brilla", "intiqa k'anchan"),
        ("la luna es hermosa", "killaqa sumaqmi"),
        ("me gusta cantar", "takiyta munani"),
        ("quiero bailar", "tusuy munani"),
        ("estoy feliz", "kusisqa kachkani"),
        ("estoy triste", "llakisqa kachkani"),
        ("te quiero", "munakuyki"),
        ("te amo", "anchata munakuyki"),
        ("es importante", "chaniyuqmi"),
        ("es necesario", "necesariom"),
    ]

    # Crear múltiples oraciones combinando las frases básicas
    spanish = []
    quechua = []

    # Agregar frases básicas
    for es, qu in data:
        spanish.append(es)
        quechua.append(qu)

    # Combinar algunas frases para crear oraciones más complejas
    for _ in range(20):
        idx1 = random.randint(0, len(data)-1)
        idx2 = random.randint(0, len(data)-1)
        spanish.append(f"{data[idx1][0]} y {data[idx2][0]}")
        quechua.append(f"{data[idx1][1]} {data[idx2][1]}")

    return spanish, quechua

# Preprocesar datos
def prepare_data():
    # Obtener datos de muestra
    spanish_sentences, quechua_sentences = create_sample_dataset()

    # Preprocesar las oraciones
    spanish_sentences, pos_tags, lemmas = preprocess_and_segment(spanish_sentences)
    quechua_sentences = [preprocess_sentence(s) for s in quechua_sentences]

    # Mostrar tabla de preprocesamiento
    preprocessing_df = pd.DataFrame({
        'Original': spanish_sentences,
        'Lemmatized': lemmas,
        'POS Tags': pos_tags
    })
    print("Tabla de Preprocesamiento:")
    print(preprocessing_df)

    # Añadir tokens especiales
    spanish_sentences = ['<start> ' + s + ' <end>' for s in spanish_sentences]
    quechua_sentences = ['<start> ' + s + ' <end>' for s in quechua_sentences]

    # Crear tokenizadores
    spanish_tokenizer = Tokenizer(filters='')
    quechua_tokenizer = Tokenizer(filters='')

    # Ajustar los tokenizadores
    spanish_tokenizer.fit_on_texts(spanish_sentences)
    quechua_tokenizer.fit_on_texts(quechua_sentences)

    # Convertir texto a secuencias
    spanish_sequences = spanish_tokenizer.texts_to_sequences(spanish_sentences)
    quechua_sequences = quechua_tokenizer.texts_to_sequences(quechua_sentences)

    # Padding de secuencias
    spanish_padded = pad_sequences(spanish_sequences, maxlen=MAX_LENGTH_SPANISH, padding='post')
    quechua_padded = pad_sequences(quechua_sequences, maxlen=MAX_LENGTH_QUECHUA, padding='post')

    # Tamaño de los vocabularios
    spanish_vocab_size = len(spanish_tokenizer.word_index) + 1  # +1 para el token 0 (padding)
    quechua_vocab_size = len(quechua_tokenizer.word_index) + 1  # +1 para el token 0 (padding)

    # Dividir en conjuntos de entrenamiento y prueba (80% - 20%)
    indices = np.arange(len(spanish_padded))
    np.random.shuffle(indices)

    train_size = int(0.8 * len(indices))

    train_indices = indices[:train_size]
    test_indices = indices[train_size:]

    x_train = spanish_padded[train_indices]
    y_train = quechua_padded[train_indices]
    x_test = spanish_padded[test_indices]
    y_test = quechua_padded[test_indices]

    # Preparar y_train y y_test para entrenamiento (one-hot encoding)
    y_train_oh = tf.keras.utils.to_categorical(y_train, num_classes=quechua_vocab_size)
    y_test_oh = tf.keras.utils.to_categorical(y_test, num_classes=quechua_vocab_size)

    return (spanish_tokenizer, quechua_tokenizer, x_train, y_train, y_train_oh,
            x_test, y_test, y_test_oh, spanish_vocab_size, quechua_vocab_size)

# Construir el modelo LSTM
def build_lstm_model(spanish_vocab_size, quechua_vocab_size):
    model = Sequential([
        # Capa de embedding para convertir índices de palabras en vectores densos
        Embedding(input_dim=spanish_vocab_size, output_dim=EMBEDDING_DIM, mask_zero=True),

        # Primera capa LSTM (codificador)
        LSTM(256, return_sequences=True),
        Dropout(0.2),

        # Segunda capa LSTM (codificador)
        LSTM(256, return_sequences=False),
        Dropout(0.2),

        # Repetir vector para la secuencia de salida
        RepeatVector(MAX_LENGTH_QUECHUA),

        # Primera capa LSTM (decodificador)
        LSTM(256, return_sequences=True),
        Dropout(0.2),

        # Segunda capa LSTM (decodificador)
        LSTM(256, return_sequences=True),
        Dropout(0.2),

        # Capa de salida
        TimeDistributed(Dense(quechua_vocab_size, activation='softmax'))
    ])

    # Compilar el modelo
    model.compile(
        optimizer='adam',
        loss='categorical_crossentropy',
        metrics=['accuracy']
    )

    return model

# Función para traducir una oración
def translate_sentence(sentence, model, spanish_tokenizer, quechua_tokenizer):
    # Preprocesar la oración
    processed_sentence = preprocess_sentence(sentence)
    processed_sentence = '<start> ' + processed_sentence + ' <end>'

    # Convertir a secuencia
    sequence = spanish_tokenizer.texts_to_sequences([processed_sentence])

    # Padding
    padded_sequence = pad_sequences(sequence, maxlen=MAX_LENGTH_SPANISH, padding='post')

    # Predecir con el modelo
    output_seq = model.predict(padded_sequence)

    # Convertir las probabilidades en índices de tokens
    output_seq = np.argmax(output_seq, axis=-1)

    # Convertir índices a palabras
    translated_words = []
    for idx in output_seq[0]:
        if idx > 0:  # Ignorar padding (0)
            word = get_word_from_index(idx, quechua_tokenizer)
            if word == '<end>':
                break
            if word != '<start>':
                translated_words.append(word)

    return ' '.join(translated_words)

def get_word_from_index(idx, tokenizer):
    """Obtiene una palabra a partir de su índice en el tokenizador"""
    for word, index in tokenizer.word_index.items():
        if index == idx:
            return word
    return ""

# Función para mostrar gráficos de entrenamiento
def plot_training_history(history):
    plt.figure(figsize=(12, 5))

    plt.subplot(1, 2, 1)
    plt.plot(history.history['loss'], label='Training Loss')
    plt.plot(history.history['val_loss'], label='Validation Loss')
    plt.title('Loss durante el entrenamiento')
    plt.xlabel('Época')
    plt.ylabel('Loss')
    plt.legend()

    plt.subplot(1, 2, 2)
    plt.plot(history.history['accuracy'], label='Training Accuracy')
    plt.plot(history.history['val_accuracy'], label='Validation Accuracy')
    plt.title('Precisión durante el entrenamiento')
    plt.xlabel('Época')
    plt.ylabel('Precisión')
    plt.legend()

    plt.tight_layout()
    plt.show()

# Función principal
def main():
    print("Iniciando proyecto de traducción español-quechua con LSTM")

    # Preparar datos
    (spanish_tokenizer, quechua_tokenizer, x_train, y_train, y_train_oh,
     x_test, y_test, y_test_oh, spanish_vocab_size, quechua_vocab_size) = prepare_data()

    print(f"Tamaño del vocabulario español: {spanish_vocab_size}")
    print(f"Tamaño del vocabulario quechua: {quechua_vocab_size}")
    print(f"Ejemplos de entrenamiento: {len(x_train)}")
    print(f"Ejemplos de prueba: {len(x_test)}")

    # Construir modelo LSTM
    print("\n=== Construyendo modelo LSTM ===")
    lstm_model = build_lstm_model(spanish_vocab_size, quechua_vocab_size)

    # Forzar la construcción del modelo con una entrada ficticia
    lstm_model(np.zeros((1, MAX_LENGTH_SPANISH)))
    lstm_model.summary()

    # Entrenar el modelo LSTM
    print("\n=== Entrenando modelo LSTM ===")
    history = lstm_model.fit(
        x_train, y_train_oh,
        batch_size=BATCH_SIZE,
        epochs=EPOCHS,
        validation_split=0.2
    )

    # Mostrar gráficos de entrenamiento
    plot_training_history(history)

    # Evaluar modelo LSTM
    loss, acc = lstm_model.evaluate(x_test, y_test_oh)
    print(f"LSTM - Loss en prueba: {loss:.4f}")
    print(f"LSTM - Precisión en prueba: {acc:.4f}")

    # Traducir algunas oraciones de prueba
    print("\n=== Ejemplos de traducción con LSTM ===")
    test_sentences = [
        "hola amigo",
        "buenos días, ¿cómo estás?",
        "me gusta la montaña",
        "estoy aprendiendo quechua",
        "el agua es vida"
    ]

    for sentence in test_sentences:
        translation = translate_sentence(sentence, lstm_model, spanish_tokenizer, quechua_tokenizer)
        print(f"ES: {sentence}")
        print(f"QU: {translation}")
        print()

    # Guardar modelo
    lstm_model.save('lstm_esp_quechua_translator.h5')
    print("\nModelo LSTM guardado con éxito.")

    print("\n=== Análisis del modelo LSTM ===")
    print("Ventajas:")
    print("- Mejor manejo de dependencias a largo plazo gracias a su arquitectura de puertas")
    print("- Soluciona el problema del desvanecimiento del gradiente")
    print("- Mayor precisión en traducciones complejas")
    print("- Capacidad para recordar información relevante por más tiempo")

    print("\nDesventajas:")
    print("- Mayor complejidad computacional y tiempo de entrenamiento")
    print("- Más parámetros que entrenar (mayor uso de memoria)")
    print("- Puede sufrir de sobreajuste en conjuntos de datos pequeños")
    print("- Mayor riesgo de explosión del gradiente sin una configuración adecuada")

    print("\nProyecto LSTM completado con éxito!")

if __name__ == "__main__":
    main()
```

```python
import numpy as np
import tensorflow as tf
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import Sequential, load_model
from tensorflow.keras.layers import Embedding, SimpleRNN, LSTM, Dense, RepeatVector, TimeDistributed, Dropout
import re
import unicodedata
import matplotlib.pyplot as plt
import random
import pandas as pd
import spacy

# Intentar cargar modelo de Spacy para español
try:
    nlp = spacy.load('es_core_news_sm')
    use_spacy = True
except:
    print("Spacy no disponible, se usará preprocesamiento básico.")
    use_spacy = False

# Constantes y configuración
MAX_LENGTH_SPANISH = 20  # Reducido para mejor rendimiento con datos limitados
MAX_LENGTH_QUECHUA = 20
EMBEDDING_DIM = 128     # Reducido para evitar sobreajuste con datos limitados
BATCH_SIZE = 32        # Reducido para mejor generalización
EPOCHS = 50            # Aumentado para mejor aprendizaje
VALIDATION_SPLIT = 0.2

# Función para normalizar texto
def unicode_to_ascii(s):
    """Normaliza caracteres unicode a ASCII pero preserva caracteres especiales quechua"""
    # No eliminamos caracteres diacríticos útiles para quechua
    return ''.join(c for c in unicodedata.normalize('NFC', s))

def preprocess_sentence(sentence):
    """Preprocesa una oración"""
    # Convertir a minúsculas y eliminar espacios al inicio/final
    sentence = sentence.lower().strip()

    # Añadir espacios alrededor de los signos de puntuación
    sentence = re.sub(r'([!.?,])', r' \1 ', sentence)

    # Eliminar caracteres no deseados pero manteniendo acentos y caracteres quechua
    sentence = re.sub(r'[^a-záéíóúüñ\'\-\.!?,\s]', ' ', sentence)

    # Eliminar espacios múltiples
    sentence = re.sub(r'\s+', ' ', sentence)

    return sentence.strip()

# Función para segmentar frases y realizar preprocesamiento adicional
def preprocess_and_segment(sentences):
    segmented_sentences = []
    pos_tags = []
    lemmas = []

    if use_spacy:
        for sentence in sentences:
            doc = nlp(sentence)
            lemmatized_sentence = ' '.join([token.lemma_ for token in doc])
            pos_tag = ' '.join([token.pos_ for token in doc])
            segmented_sentences.append(lemmatized_sentence)
            pos_tags.append(pos_tag)
            lemmas.append([token.lemma_ for token in doc])
    else:
        # Preprocesamiento básico si no hay spaCy
        for sentence in sentences:
            processed = preprocess_sentence(sentence)
            segmented_sentences.append(processed)
            pos_tags.append("")
            lemmas.append(processed.split())

    return segmented_sentences, pos_tags, lemmas

# Crear dataset ampliado de muestra
def create_sample_dataset():
    """Crea un conjunto de datos de muestra ampliado para entrenamiento"""
    # Datos básicos - pares español-quechua
    basic_data = [
        ("buenos días", "allin p'unchaw"),
        ("¿cómo estás?", "imaynalla kashanki?"),
        ("estoy bien", "allillanmi kashani"),
        ("me llamo Juan", "ñuqaq sutinmi Juan"),
        ("¿cuál es tu nombre?", "imayna sutin kanki?"),
        ("nos vemos", "tupananchiskama"),
        ("hasta mañana", "paqariykama"),
        ("¿dónde está la plaza?", "maypita plaza kachkan?"),
        ("necesito ayuda", "yanapaynita munani"),
        ("quiero cantar", "takiynita munani"),
        ("¿hablas quechua?", "runasimi rimanqachu?"),
        ("estoy aprendiendo quechua", "runasimi yachachkani"),
        ("eres mi amigo", "qam ñuqapa mayllayki kanki"),
        ("el cielo es azul", "hanaq pacha anqasmi"),
        ("la montaña es grande", "urqu hatunmi"),
        ("el agua es vida", "unu kawsaymi"),
        ("el sol brilla", "inti k'anchakuchkan"),
        ("la luna es hermosa", "killa sumaqmi"),
        ("me gusta cantar", "takiynita munani"),
        ("quiero bailar", "tusuyta munani"),
        ("estoy feliz", "kusisqam kashani"),
        ("estoy triste", "llakisqam kashani"),
        ("te quiero", "munayki"),
        ("te amo", "waylluyki"),
        ("es importante", "chaniyuqmi"),
        ("es necesario", "kananmi"),
    ]

    # Crear frases básicas
    spanish = []
    quechua = []

    # Agregar frases básicas
    for es, qu in basic_data:
        spanish.append(es)
        quechua.append(qu)

    # Combinar algunas frases para crear oraciones más complejas
    for _ in range(200):  # Aumentado a 200 combinaciones para más datos
        idx1 = random.randint(0, len(basic_data)-1)
        idx2 = random.randint(0, len(basic_data)-1)
        spanish.append(f"{basic_data[idx1][0]} y {basic_data[idx2][0]}")
        quechua.append(f"{basic_data[idx1][1]} {basic_data[idx2][1]}")

    # Crear algunas frases con "me gusta X"
    things = ["casa", "comida", "música", "libro", "perro", "gato", "agua", "sol", "luna", "río"]
    things_quechua = ["wasi", "mikhuy", "taki", "qillqa", "allqu", "michi", "unu", "inti", "killa", "mayu"]

    for i in range(len(things)):
        spanish.append(f"me gusta la {things[i]}")
        quechua.append(f"{things_quechua[i]}ta munani")

        spanish.append(f"no me gusta la {things[i]}")
        quechua.append(f"mana {things_quechua[i]}ta munanikhu")

    # Crear algunas frases de posesión
    possessions = ["casa", "perro", "libro", "auto", "familia"]
    possessions_quechua = ["wasi", "allqu", "qillqa", "carro", "ayllu"]

    for i in range(len(possessions)):
        spanish.append(f"tengo una {possessions[i]}")
        quechua.append(f"{possessions_quechua[i]}ywanmi kani")

        spanish.append(f"no tengo {possessions[i]}")
        quechua.append(f"manan {possessions_quechua[i]}ywanchu kani")

    return spanish, quechua

# Preprocesar datos
def prepare_data():
    # Obtener datos de muestra
    spanish_sentences, quechua_sentences = create_sample_dataset()

    print(f"Tamaño del dataset: {len(spanish_sentences)} pares de oraciones")

    # Preprocesar las oraciones
    spanish_clean = [preprocess_sentence(s) for s in spanish_sentences]
    spanish_sentences, _, _ = preprocess_and_segment(spanish_clean)
    quechua_sentences = [preprocess_sentence(s) for s in quechua_sentences]

    # Mostrar tabla de preprocesamiento para algunas oraciones
    sample_indices = np.random.choice(len(spanish_sentences), min(10, len(spanish_sentences)), replace=False)
    preprocessing_samples = []

    for idx in sample_indices:
        preprocessing_samples.append({
            'Original': spanish_sentences[idx],
            'Quechua': quechua_sentences[idx]
        })

    preprocessing_df = pd.DataFrame(preprocessing_samples)
    print("Muestra de datos de entrenamiento:")
    print(preprocessing_df)

    # Añadir tokens especiales
    spanish_with_tokens = ['<start> ' + s + ' <end>' for s in spanish_sentences]
    quechua_with_tokens = ['<start> ' + s + ' <end>' for s in quechua_sentences]

    # Crear tokenizadores
    spanish_tokenizer = Tokenizer(filters='', oov_token='<unk>')
    quechua_tokenizer = Tokenizer(filters='', oov_token='<unk>')

    # Ajustar los tokenizadores
    spanish_tokenizer.fit_on_texts(spanish_with_tokens)
    quechua_tokenizer.fit_on_texts(quechua_with_tokens)

    # Convertir texto a secuencias
    spanish_sequences = spanish_tokenizer.texts_to_sequences(spanish_with_tokens)
    quechua_sequences = quechua_tokenizer.texts_to_sequences(quechua_with_tokens)

    # Padding de secuencias
    spanish_padded = pad_sequences(spanish_sequences, maxlen=MAX_LENGTH_SPANISH, padding='post')
    quechua_padded = pad_sequences(quechua_sequences, maxlen=MAX_LENGTH_QUECHUA, padding='post')

    # Tamaño de los vocabularios
    spanish_vocab_size = len(spanish_tokenizer.word_index) + 1  # +1 para el token 0 (padding)
    quechua_vocab_size = len(quechua_tokenizer.word_index) + 1  # +1 para el token 0 (padding)

    # Dividir en conjuntos de entrenamiento y prueba (80% - 20%)
    indices = np.arange(len(spanish_padded))
    np.random.shuffle(indices)

    train_size = int(0.8 * len(indices))

    train_indices = indices[:train_size]
    test_indices = indices[train_size:]

    x_train = spanish_padded[train_indices]
    y_train = quechua_padded[train_indices]
    x_test = spanish_padded[test_indices]
    y_test = quechua_padded[test_indices]

    # Preparar y_train y y_test para entrenamiento (one-hot encoding)
    y_train_oh = tf.keras.utils.to_categorical(y_train, num_classes=quechua_vocab_size)
    y_test_oh = tf.keras.utils.to_categorical(y_test, num_classes=quechua_vocab_size)

    # Imprimir algunas estadísticas
    print(f"Ejemplos de entrenamiento: {len(x_train)}")
    print(f"Ejemplos de prueba: {len(x_test)}")
    print(f"Tamaño del vocabulario español: {spanish_vocab_size}")
    print(f"Tamaño del vocabulario quechua: {quechua_vocab_size}")

    # Mostrar las primeras palabras de cada vocabulario
    print("\nMuestra del vocabulario español:")
    word_idx = list(spanish_tokenizer.word_index.items())[:10]
    for word, idx in word_idx:
        print(f"{word}: {idx}")

    print("\nMuestra del vocabulario quechua:")
    word_idx = list(quechua_tokenizer.word_index.items())[:10]
    for word, idx in word_idx:
        print(f"{word}: {idx}")

    return (spanish_tokenizer, quechua_tokenizer, x_train, y_train, y_train_oh,
            x_test, y_test, y_test_oh, spanish_vocab_size, quechua_vocab_size)

# Construir el modelo RNN simple
def build_rnn_model(spanish_vocab_size, quechua_vocab_size):
    model = Sequential([
        # Capa de embedding para convertir índices de palabras en vectores densos
        Embedding(input_dim=spanish_vocab_size, output_dim=EMBEDDING_DIM, mask_zero=True),

        # Capa RNN simple para codificar la secuencia de entrada
        SimpleRNN(256, return_sequences=False),

        # Capa de dropout para regularización
        Dropout(0.3),  # Aumentado para evitar sobreajuste

        # Repetir el vector para la secuencia de salida
        RepeatVector(MAX_LENGTH_QUECHUA),

        # RNN decodificador que genera la secuencia de salida
        SimpleRNN(256, return_sequences=True),

        # Capa de dropout
        Dropout(0.3),  # Aumentado para evitar sobreajuste

        # Capa de salida (una distribución de probabilidad sobre el vocabulario quechua para cada posición)
        TimeDistributed(Dense(quechua_vocab_size, activation='softmax'))
    ])

    # Compilar el modelo
    model.compile(
        optimizer=tf.keras.optimizers.Adam(learning_rate=0.001),  # Tasa de aprendizaje explícita
        loss='categorical_crossentropy',
        metrics=['accuracy']
    )

    return model

# Construir el modelo LSTM
def build_lstm_model(spanish_vocab_size, quechua_vocab_size):
    model = Sequential([
        # Capa de embedding para convertir índices de palabras en vectores densos
        Embedding(input_dim=spanish_vocab_size, output_dim=EMBEDDING_DIM, mask_zero=True),

        # Primera capa LSTM (codificador)
        LSTM(256, return_sequences=True),
        Dropout(0.3),  # Aumentado para evitar sobreajuste

        # Segunda capa LSTM (codificador)
        LSTM(256, return_sequences=False),
        Dropout(0.3),  # Aumentado para evitar sobreajuste

        # Repetir vector para la secuencia de salida
        RepeatVector(MAX_LENGTH_QUECHUA),

        # Primera capa LSTM (decodificador)
        LSTM(256, return_sequences=True),
        Dropout(0.3),  # Aumentado para evitar sobreajuste

        # Segunda capa LSTM (decodificador)
        LSTM(256, return_sequences=True),
        Dropout(0.3),  # Aumentado para evitar sobreajuste

        # Capa de salida
        TimeDistributed(Dense(quechua_vocab_size, activation='softmax'))
    ])

    # Compilar el modelo
    model.compile(
        optimizer=tf.keras.optimizers.Adam(learning_rate=0.001),  # Tasa de aprendizaje explícita
        loss='categorical_crossentropy',
        metrics=['accuracy']
    )

    return model

# Función para traducir una oración - CORREGIDA
def translate_sentence(sentence, model, spanish_tokenizer, quechua_tokenizer):
    # Preprocesar la oración
    processed_sentence = preprocess_sentence(sentence)

    # Aplicar procesamiento adicional si está disponible spaCy
    if use_spacy:
        doc = nlp(processed_sentence)
        processed_sentence = ' '.join([token.lemma_ for token in doc])

    # Añadir tokens especiales
    processed_sentence = '<start> ' + processed_sentence + ' <end>'

    # Convertir a secuencia
    sequence = spanish_tokenizer.texts_to_sequences([processed_sentence])

    # Padding
    padded_sequence = pad_sequences(sequence, maxlen=MAX_LENGTH_SPANISH, padding='post')

    # Predecir con el modelo
    predictions = model.predict(padded_sequence)[0]

    # Convertir las probabilidades en índices de tokens
    predicted_indices = np.argmax(predictions, axis=1)

    # Convertir índices a palabras
    translated_words = []
    # Buscar el índice del token de fin
    end_token = quechua_tokenizer.word_index.get('<end>', 0)

    for idx in predicted_indices:
        if idx == 0:  # Ignorar padding
            continue

        # Obtener la palabra correspondiente al índice
        word = None
        for w, i in quechua_tokenizer.word_index.items():
            if i == idx:
                word = w
                break

        if word is None:
            continue
        if word == '<end>':
            break
        if word == '<start>':
            continue

        translated_words.append(word)

    # Unir palabras y devolver traducción
    return ' '.join(translated_words)

# Función para mostrar gráficos de entrenamiento
def plot_training_history(history, model_name):
    plt.figure(figsize=(12, 5))

    # Gráfico de pérdida (loss)
    plt.subplot(1, 2, 1)
    plt.plot(history.history['loss'], label='Training Loss')
    plt.plot(history.history['val_loss'], label='Validation Loss')
    plt.title(f'{model_name} - Loss')
    plt.xlabel('Epochs')
    plt.ylabel('Loss')
    plt.legend()

    # Gráfico de precisión (accuracy)
    plt.subplot(1, 2, 2)
    plt.plot(history.history['accuracy'], label='Training Accuracy')
    plt.plot(history.history['val_accuracy'], label='Validation Accuracy')
    plt.title(f'{model_name} - Accuracy')
    plt.xlabel('Epochs')
    plt.ylabel('Accuracy')
    plt.legend()

    plt.tight_layout()
    plt.show()

# Función para entrenar y evaluar el modelo
def train_and_evaluate_model(model, x_train, y_train_oh, x_test, y_test_oh, model_name):
    # Entrenar el modelo
    history = model.fit(
        x_train, y_train_oh,
        epochs=EPOCHS,
        batch_size=BATCH_SIZE,
        validation_split=VALIDATION_SPLIT,
        verbose=1
    )

    # Evaluar el modelo
    test_loss, test_accuracy = model.evaluate(x_test, y_test_oh, verbose=1)
    print(f"{model_name} - Test Loss: {test_loss:.4f}")
    print(f"{model_name} - Test Accuracy: {test_accuracy:.4f}")

    # Guardar el modelo
    model.save(f"{model_name}.h5")

    # Mostrar gráficos de entrenamiento
    plot_training_history(history, model_name)

    return history

# Ejecutar el flujo completo
if __name__ == "__main__":
    # Preparar datos
    (spanish_tokenizer, quechua_tokenizer, x_train, y_train, y_train_oh,
     x_test, y_test, y_test_oh, spanish_vocab_size, quechua_vocab_size) = prepare_data()

    # Construir y entrenar el modelo RNN
    rnn_model = build_rnn_model(spanish_vocab_size, quechua_vocab_size)
    print("Entrenando modelo RNN...")
    rnn_history = train_and_evaluate_model(rnn_model, x_train, y_train_oh, x_test, y_test_oh, "RNN")

    # Construir y entrenar el modelo LSTM
    lstm_model = build_lstm_model(spanish_vocab_size, quechua_vocab_size)
    print("Entrenando modelo LSTM...")
    lstm_history = train_and_evaluate_model(lstm_model, x_train, y_train_oh, x_test, y_test_oh, "LSTM")

    # Probar la traducción con algunas oraciones
    test_sentences = [
        "hola, ¿cómo estás?",
        "me gusta la música",
        "necesito ayuda",
        "el cielo es azul",
        "quiero aprender quechua"
    ]

    print("\nTraducciones con el modelo RNN:")
    for sentence in test_sentences:
        translation = translate_sentence(sentence, rnn_model, spanish_tokenizer, quechua_tokenizer)
        print(f"{sentence} -> {translation}")

    print("\nTraducciones con el modelo LSTM:")
    for sentence in test_sentences:
        translation = translate_sentence(sentence, lstm_model, spanish_tokenizer, quechua_tokenizer)
        print(f"{sentence} -> {translation}")

    # Interfaz de usuario para traducción
    print("\n=== Interfaz de usuario para traducción ===")
    while True:
        user_input = input("Ingrese una oración en español (o escriba 'salir' para terminar): ")
        if user_input.lower() == 'salir':
            break

        print("Traduciendo...")
        rnn_translation = translate_sentence(user_input, rnn_model, spanish_tokenizer, quechua_tokenizer)
        lstm_translation = translate_sentence(user_input, lstm_model, spanish_tokenizer, quechua_tokenizer)
        print(f"Traducción (RNN): {rnn_translation}")
        print(f"Traducción (LSTM): {lstm_translation}")
        print()

    print("\nProyecto completado con éxito!")
```
